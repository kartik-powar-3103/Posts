# Loss Functions

```python
import numpy as np
```

```python
# Squared Loss 
def squared_loss_fn(z):
    return z**2
```

```python
# Absolute Loss 
def absolute_loss_fn(z):
    return abs(z)
```

```python
# Huber Loss
def huber_loss_fn(z, delta):
    return np.where(
        np.abs(z) <= delta, 
        (1/2) * squared_loss_fn(z),
        delta * (absolute_loss_fn(z) - (delta/2))
    )
```

```python

# Log Cosh Loss
def cosh_fn(z):
    return (np.exp(z) + np.exp(-1*z))/2

def log_cosh_loss_fn(z):
    return np.log(cosh_fn(z))
```

```python
import matplotlib.pyplot as plt 
```

```python
print("Huber Loss (\u2202 = 1)")
```

```python
# define true values
y_true = 0

# generate prediction around true value
y_pred  = np.linspace(-2, 2, 100)

# compute loss values
z = (y_true - y_pred) 

squared_loss_values = squared_loss_fn(z)
absolute_loss_values = absolute_loss_fn(z)
huber_loss_values_d1 = huber_loss_fn(z, delta=1)
log_cosh_loss_values = log_cosh_loss_fn(z)

y_pred_d5  = np.linspace(-2, 2, 100)
z_d5 = (y_true - y_pred_d5) 
huber_loss_values_d5 = huber_loss_fn(z_d5, delta=5)


# visualizing the loss
plt.figure(figsize=(7,5))
plt.plot(y_pred, squared_loss_values, label = 'Squared Loss', color = 'b')
plt.plot(y_pred, absolute_loss_values, label = 'Absolute Loss', color = 'y')
plt.plot(y_pred, huber_loss_values_d1, label = 'Huber Loss (\u2202 =  1)', color = 'g')
plt.plot(y_pred, huber_loss_values_d5, label = 'Huber Loss (\u2202 =  5)', color = 'm')
plt.plot(y_pred, log_cosh_loss_values, label = 'Log Cosh Loss', color = 'c')
plt.xlabel(r"Predicted Value ($\hat{y}$)")
plt.ylabel("Loss Values")
plt.axvline(y_true, linestyle="-", color="red", label="True Value (y)")
plt.legend()
plt.grid()
plt.show()
```
